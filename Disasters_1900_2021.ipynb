```
px.area(df.groupby('Year').count(),df.groupby('Year').count().index, 'Disaster Group', title= 'Disaster Count Per Year', template='presentation', width=800).update_layout(xaxis_title = 'Year', yaxis_title = 'Disaster Count')
```

* **532 disasters occured in year-2002**

```
px.line(df.groupby('Year').sum(),df.groupby('Year').count().index, 'Total Affected', title= 'Disaster Affect by Year', template='presentation', width=800).update_layout(xaxis_title = 'Year', yaxis_title = 'No. of Affected People')
```

* **658 M people are affected in 2002 disasters**

```
px.pie(df, 'Continent',labels='Continent', title= 'Disasters by Continent', template='presentation', width=500).update_traces(textinfo='label+percent+value', textposition='inside')
```

* We have null value in totaldeaths column but we can't remove or fill that column and for some specific task we have to replace null value with 0.

```
df['Total Deaths'].replace(np.NaN, 0, inplace = True)
#fillna()
```

```
px.scatter(df.sort_values('Total Deaths', ascending = False).head(30), y ='Total Deaths', x = 'Country', color='Continent', hover_data=df.columns, size='Total Deaths', size_max=30, title = 'Top Disasters With Highest No. Deaths', template='presentation', width=900)
```
